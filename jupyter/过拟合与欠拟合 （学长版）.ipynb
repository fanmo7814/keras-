```python
import keras
from keras import layers
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#信用卡欺诈数据集
```

```python
data=pd.read_csv('./credit-a.csv',header=None)   
#因为第一行就是数据 所以要columns来加一个1234来整理
```

```python
model=keras.Sequential()
```

```python
model.summary()
```

```python
y   #改变y    让其为0和1   而不是-1和1
```

```python
history=model.fit(x,y,epochs=1000)
```

```python
plt.plot(range(1000),history.history.get('acc'),c='r')
plt.plot(history.epoch,history.history.get('loss'),c='b')  #画一起
```

评价标准 ： 对未知数据的预测能力

```python
x_train=x[:int(len(x)*0.75)]     #x总长的75%
x_test=x[int(len(x)*0.75):]     #注意算式的写法！
y_train=y[:int(len(y)*0.75)]
y_test= y[int(len(y)*0.75):]                        #75%做训练数据   25%做测试数据
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
model=keras.Sequential()
```

```python
from keras import layers
```

```python
model.add(layers.Dense(128, input_dim=15,activation='relu'))
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer='adam',
             loss='binary_crossentropy',
             metrics=['acc'])
```

```python
history=model.fit(x_train,y_train,epochs=1000,validation_data=(x_test,y_test))
                                              #将测试数据的变化率也显示
```

```python
plt.plot(history.epoch,history.history.get('acc'),c='b',label='acc')
plt.plot(history.epoch,history.history.get('val_acc'),c='r',label='val_acc')
plt.legend()    #记得要加legend函数  才可右下角有注释
```

过拟合：训练数据和测试数据 得到的有比较大的差距 “过”  实际上就是
真实的比不上训练的

在训练上的数据   得到最后的98%的正确率   
现在到test数据上进行检验

```python
model.evaluate(x_train,y_train)   #在训练数据上的检验    答案前一个数据为loss  
                                                        #后一个为准确率
```

```python
model.evaluate(x_test,y_test)   #在测试数据上的检验
```

Dropout抑制过拟合原理：


随机丢弃一些隐藏单元（隐含层）
来增强剩下几个单元的机器训练
从而提高准确率

```python
model=keras.Sequential()
model.add(layers.Dense(128, input_dim=15,activation='relu'))
model.add(layers.Dropout(0.5))   #每一层之后添加一个dropout层 里面的参数为比例  
                                #随机舍弃50%的神经单元
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',
             loss='binary_crossentropy',
             metrics=['acc'])
```

```python
history=model.fit(x_train,y_train,epochs=1000,validation_data=(x_test,y_test))
```

```python
model.evaluate(x_train,y_train)
```

```python
model.evaluate(x_test,y_test)
```

```python
plt.plot(history.epoch,history.history.get('acc'),c='b',label='acc')
plt.plot(history.epoch,history.history.get('val_acc'),c='r',label='val_acc')
plt.legend()    #记得要加legend函数  才可右下角有注释
```

    总的来说 训练和测试数据靠的很近  差距明显减小   在添加了dropout层之后
    重点是在 test数据上得到了更高的准确率
           虽然在train数据准确率降低   但是最重要的是在未知数据上的准确性

/////////使用正则化抑制过拟合///////////

l1 
loss=s*abs(w1+w2+...)+mse

l2  
loss=s*(w1^2+w2^2+..)+mse        #下面按 shift+tab  可以看隐含信息

```python
from keras import regularizers
```

```python
model=keras.Sequential()
model.add(layers.Dense(128, kernel_regularizer=regularizers.l2(0.005),input_dim=15,activation='relu'))
model.add(layers.Dense(128,kernel_regularizer=regularizers.l2(0.005), activation='relu'))      #抑制过拟合  l2正则更好（是L 2）
model.add(layers.Dense(128,kernel_regularizer=regularizers.l2(0.005), activation='relu'))
model.add(layers.Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer='adam',
             loss='binary_crossentropy',
             metrics=['acc'])
```

```python
model.summary()
```

```python
history=model.fit(x_train,y_train,epochs=1000,validation_data=(x_test,y_test))
```

```python
model.evaluate(x_train,y_train)
```

```python
model.evaluate(x_test,y_test)    #谁在测试数据上得到高分就用哪个
```

```python
plt.plot(history.epoch,history.history.get('acc'),c='b',label='acc')
plt.plot(history.epoch,history.history.get('val_acc'),c='r',label='val_acc')
plt.legend()
```

虽然这次的训练没有之前的dropout效果好 但并不是函数不行 
而是  超参数的选择为0.005  有更好的 需要自己去调试  这也是机器学习的特点
和深度学习的一个   比较有争议性的  没有好的公式或者标准去用到特定的式子
还是在靠经验积累

网络参数选择的总原则：：
1：什么是神经网络的容量： 可以把可训练参数当作神经网络的容量   例本题的可训练参数
 Total params 为35201     总量越大  神经网络拟合能力越强   


```python
#欠拟合  ：  对数据的拟合能力不够 比如练到一半 准确率不变了  不能再进一步学习了
#因为其网络容量不足以支持
```

构建网络的总原则
一：增大网络容量，直到过拟合
二：采取措施抑制过拟合
三：继续增大网络容量，知道过拟合
