```python
# 编译模型
model.compile(optimizer='adam',             # 优化函数,优化算法
              loss='mse'                 # 损失函数，均方差
)
```

```python
# 训练模型
model.fit(x, y, epochs=3000)            # epochs把所以数据训练3000遍
```

```python
import pandas as pd  # pandas数据预处理工具
```

```python
data = pd.read_csv('./Advertising.csv')  #导入数据集
```

```python
data.head()  #读取前五行
```

```python
x = data[data.columns[1: -1]]  #x快速取值
```

```python
y = data.iloc[:, -1] #y取最后一列
```

```python
import tensorflow as tf
from tensorflow import keras   # 引用层
model = keras.Sequential()    # 初始化顺序模型
from tensorflow.keras import layers #从keras引用层
```

```python
model.add(layers.Dense(1, input_dim=3)) # y_pred = w1*t1 + w2*t2 + w3*t3 + b
```

```python
model.summary()  #显示有多少个参数
```

```python
model.compile(optimizer='adam',  #编译模型，使用adam算法优化loss,mes表示均方差
              loss='mse'
)
```

```python
model.fit(x, y, epochs=2000)  #训练2000次
```

```python
model.predict(pd.DataFrame([[300, 0, 0]]))   #预测
```
